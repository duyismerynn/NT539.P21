Chọn hàng bất kỳ theo giá trị một cột: hàng đầu tiên có giá trị port đích là 80

```python
df_filtered = df[df[' Destination Port'] == 80]
df_filtered.head(1)
```

Xác định số lượng giá trị Null, Nan, Inf trong dataset, chuyển chúng thành giá trị 0

```python
null_count = df.isnull().sum().sum()
nan_count = df.isna().sum().sum()
inf_count = np.isinf(df.select_dtypes(include=np.number)).sum().sum()

print('Số lượng giá trị Null: ', f"{null_count}")
print('Số lượng giá trị Nan: ', f"{nan_count}")
print('Số lượng giá trị Inf: ', f"{inf_count}")

df = df.replace([np.nan, np.inf, -np.inf], 0)
```

Hiển thị các dòng có nhãn khác “BENIGN”

```python
df_filtered = df[df['Label'] != 'BENIGN']
df_filtered.head(5)
```

Vẽ biểu đồ thể hiện phân phối theo giá trị FlowID

```python
import matplotlib.pyplot as plt

flow_id = df['Flow ID']
flow_id_counts = flow_id.value_counts()

flow_id_top10 = flow_id_counts.head(10)

plt.figure(figsize=(14, 6))
plt.bar(flow_id_top10.index, flow_id_top10.values)
plt.title('Phân phối theo giá trị FlowID')
plt.xlabel('Flow ID')
plt.ylabel('Số lượng')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
